```python
def list_toString(df, prefix):
    df = df.applymap(lambda x: ', '.join(map(str, x)) if isinstance(x, list) else x)
    df = df.applymap(lambda x: str(x).replace(prefix, '').strip() if isinstance(x, str) else x)
    return df

def dic_toString(df, prefix):
    df = df.applymap(lambda x: ', '.join([f'{k}: {v}' for k, v in x.items()]) if isinstance(x, dict) else x)
    df = df.applymap(lambda x: str(x).replace(prefix, '').strip() if isinstance(x, str) else x)
    return df
```
